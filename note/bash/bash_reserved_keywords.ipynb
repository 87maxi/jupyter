## Parte 1: Tipos de Datos Fundamentales

Bash no tiene un sistema de tipos estricto como otros lenguajes. Fundamentalmente, **todo en Bash es una cadena de texto (string)**. Sin embargo, el shell puede interpretar estas cadenas como nĂşmeros o arreglos segĂşn el contexto y los operadores utilizados.

### Cadenas de Texto (Strings)
El tipo de dato principal. Se pueden definir con o sin comillas. Las comillas dobles (`""`) permiten la expansiĂłn de variables, mientras que las comillas simples (`''`) no.

```bash
#!/bin/bash
USUARIO="Maxi"

# Comillas dobles (expansiĂłn)
echo "Hola, $USUARIO"

# Comillas simples (literal)
echo 'Hola, $USUARIO'
```

### NĂşmeros (tratados como Cadenas)
Aunque se almacenan como cadenas, Bash puede realizar operaciones aritmĂŠticas enteras mediante contextos especiales como `$((...))` o `declare -i`.

```bash
#!/bin/bash
A=10
B=20
SUMA=$((A + B))
echo "La suma es: $SUMA"

# Declarar como entero
declare -i C=0
C=C+5
echo "Contador: $C"
```

### Arreglos (Arrays)
Bash soporta dos tipos de arreglos:
1.  **Indexados**: Una lista de cadenas accedida por un Ă­ndice numĂŠrico (empezando en 0).
2.  **Asociativos**: Un conjunto de pares clave-valor (hash map), disponible desde Bash 4.0+. Requiere `declare -A`.

```bash
#!/bin/bash
# Array indexado
servidores=("web01" "db01" "cache01")
echo "Primer servidor: ${servidores[0]}"
echo "Todos: ${servidores[@]}"

# Array asociativo
declare -A puertos
puertos[http]=80
puertos[https]=443
echo "Puerto HTTPS: ${puertos[https]}"
```

## Parte 2: Referencia Exhaustiva de Palabras Reservadas

Las palabras se listan en orden alfabĂŠtico para una consulta rĂĄpida.

### `break`
Sale inmediatamente del bucle mĂĄs interno (`for`, `while`, `until` o `select`).

### `case` / `in` / `esac`
Estructura de control para mĂşltiples condiciones basada en la coincidencia de patrones, similar a un `switch`.

```bash
#!/bin/bash
RESPUESTA="si"
case $RESPUESTA in
  si|s|S)
    echo "Afirmativo"
    ;;
  no|n|N)
    echo "Negativo"
    ;;
  *)
    echo "Respuesta no reconocida"
    ;;
esac
```

### `continue`
Salta el resto de la iteraciĂłn actual de un bucle y pasa a la siguiente.

### `declare` / `typeset`
Declara variables y/o les asigna atributos. `typeset` es un alias.

### `do` / `done`
Delimitan el cuerpo de un bucle (`for`, `while`, `until`, `select`).

### `elif`
Forma parte de una estructura `if` para aĂąadir una condiciĂłn alternativa ('else if').

### `else`
Parte de una estructura `if` o `case` que se ejecuta si ninguna condiciĂłn anterior es verdadera.

### `fi`
Cierra un bloque `if` (es `if` al revĂŠs).

### `for`
Inicia un bucle que itera sobre una lista de elementos o sigue una sintaxis de estilo C.

### `function`
Palabra clave opcional para declarar una funciĂłn.

### `if` / `then`
Inicia una estructura de control condicional. `then` separa la condiciĂłn del cuerpo.

### `select`
Crea un menĂş interactivo a partir de una lista de elementos.

### `time`
Mide el tiempo de ejecuciĂłn de un comando o una tuberĂ­a. No es para control de flujo, pero es una palabra reservada.

```bash
time sleep 1
```

### `until`
Inicia un bucle que se ejecuta mientras una condiciĂłn sea falsa.

### `while`
Inicia un bucle que se ejecuta mientras una condiciĂłn sea verdadera.

### `{` / `}`
Agrupa comandos para que se ejecuten en el shell actual. A diferencia de `( )`, no crea una sub-shell. TambiĂŠn se usa para delimitar cuerpos de funciones.

```bash
mi_funcion() { echo "Cuerpo de la funciĂłn"; }
```

### `[[` / `]]`
Inicia una expresiĂłn condicional extendida, mĂĄs potente y segura que el comando `[`.

### `((` / `))`
Crea un contexto aritmĂŠtico para evaluar expresiones enteras con una sintaxis similar a C.

### `!`
Niega el estado de salida de un comando o una tuberĂ­a. TambiĂŠn se usa para la expansiĂłn del historial.

```bash
if ! grep -q "usuario" /etc/passwd; then
    echo "El usuario no existe"
fi
```
