# Referencia Completa de Sintaxis Bash

Bash (Bourne Again SHell) es el intérprete de comandos estándar en la mayoría de los sistemas Linux. Este notebook sirve como una guía detallada de su sintaxis y capacidades para el scripting.

## 1. Variables y Tipos de Datos

En Bash, no hay tipos de datos estrictos. Todo se trata esencialmente como una cadena de texto, aunque se pueden realizar operaciones aritméticas.

```bash
#!/bin/bash

# 1. Definición (Sin espacios alrededor del =)
nombre="Usuario"
edad=30

# 2. Uso de variables ($)
echo "Hola $nombre, tienes $edad años"

# 3. Variables de solo lectura
readonly PI=3.14159

# 4. Variables de entorno (export)
export MI_APP_ENV="produccion"

# 5. Capturar salida de comando (Command Substitution)
fecha=$(date +%Y-%m-%d)
usuarios_activos=$(who | wc -l)

echo "Hoy es $fecha y hay $usuarios_activos usuarios conectados"
```

## 2. Operaciones Aritméticas

Se utiliza la sintaxis `$(( ))` para realizar cálculos enteros.

```bash
num1=10
num2=5

suma=$((num1 + num2))
multiplicacion=$((num1 * num2))
division=$((num1 / num2))

echo "Suma: $suma"

# Para operaciones con decimales (flotantes), se usa bc (Binary Calculator)
resultado_float=$(echo "scale=2; 10 / 3" | bc)
echo "División decimal: $resultado_float"
```

## 3. Estructuras de Control

### Condicionales (if)
Es vital usar `[[ ]]` (bashismo preferido) para las pruebas, ya que es más potente que `[ ]`.

```bash
numero=15

if [[ $numero -gt 10 ]]; then
    echo "El número es mayor a 10"
elif [[ $numero -eq 10 ]]; then
    echo "El número es exactamente 10"
else
    echo "El número es menor a 10"
fi

# Comparación de strings
str="hola"
if [[ $str == "hola" ]]; then
    echo "Coincide"
fi
```

### Bucles (for, while)
Bash ofrece varias formas de iterar.

```bash
# For estilo lista
for fruta in manzana naranja platano; do
    echo "Me gusta la $fruta"
done

# For estilo C
for ((i=0; i<3; i++)); do
    echo "Contador: $i"
done

# While
contador=3
while [[ $contador -gt 0 ]]; do
    echo "Quedan $contador"
    ((contador--))
done
```

## 4. Arrays (Arreglos)

Bash soporta arreglos indexados y asociativos (desde la versión 4.0).

```bash
# Array indexado
servidores=("web01" "db01" "cache01")
echo "Primer servidor: ${servidores[0]}"
echo "Todos: ${servidores[@]}"
echo "Total: ${#servidores[@]}"

# Array asociativo
declare -A configuracion
configuracion[puerto]=8080
configuracion[host]="localhost"

echo "Host: ${configuracion[host]}"
```

## 5. Funciones

Las funciones en Bash no reciben parámetros nombrados; usan variables posicionales ($1, $2, etc.).

```bash
saludar() {
    local nombre=$1  # Usar local para no contaminar el scope global
    echo "Hola $nombre"
    return 0         # Status de salida (0-255)
}

saludar "Maxi"

# Capturar el status de la última ejecución ($?)
if [[ $? -eq 0 ]]; then
    echo "Comando ejecutado con éxito"
fi
```

## 6. Manejo de Argumentos del Script

- `$0`: Nombre del script.
- `$1` - `$9`: Argumentos posicionales.
- `$#`: Número de argumentos.
- `$@`: Todos los argumentos como una lista.
- `$$`: PID del script actual.

## 7. Redirecciones y Pipes

- `>`: Sobrescribir archivo.
- `>>`: Añadir al archivo.
- `2>`: Redirigir errores (stderr).
- `&>`: Redirigir stdout y stderr.
- `|`: Tubería (pasa salida de un comando a otro).

## 8. Expansión y Globbing

Bash realiza varios tipos de expansiones antes de ejecutar un comando, lo que permite manipular texto y archivos de forma muy eficiente.

### Expansión de Llaves y Parámetros

```bash
# Expansión de llaves { }
echo archivo_{1,2,3}.txt  # archivo_1.txt archivo_2.txt archivo_3.txt
echo {0..10..2}           # 0 2 4 6 8 10 (Rango con incremento)

# Parameter Expansion (Manipulación de variables)
ruta="/var/log/nginx/access.log"
echo "Nombre base: ${ruta##*/}"    # Quita el prefijo más largo (access.log)
echo "Directorio: ${ruta%/*}"      # Quita el sufijo más corto (/var/log/nginx)

texto="Hola Mundo"
echo "${texto/Mundo/Bash}"        # Reemplazo: Hola Bash
echo "Longitud: ${#texto}"         # 10

# Globbing (Patrones de archivos)
# *.txt -> coincidencia múltiple
# ?.txt -> coincidencia de un solo carácter
# [a-z].txt -> coincidencia de rango
```
