```python
def fibonacci_of(n):
     if n in {0, 1}:  # Base case
         return n
     return fibonacci_of(n - 1) + fibonacci_of(n - 2)  # Recursive cas
```

```python
[ fibonacci_of(n) for n in range(15)   ]
```
