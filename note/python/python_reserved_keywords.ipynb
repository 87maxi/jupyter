## Parte 1: Tipos de Datos Fundamentales

Python es un lenguaje de **tipado dinámico**, lo que significa que el tipo de una variable se determina en tiempo de ejecución. Sin embargo, cada valor tiene un tipo bien definido.

### Tipos Numéricos

```python
# int: Enteros de precisión arbitraria.
entero = 100

# float: Números de punto flotante de doble precisión (64 bits).
flotante = 3.14

# complex: Números complejos.
complejo = 2 + 3j
```

### Secuencias
Colecciones ordenadas de elementos.

```python
# str: Cadenas de texto inmutables.
cadena = "Hola, Python"

# list: Secuencias mutables y ordenadas de objetos.
lista = [1, "a", 3.0, True]

# tuple: Secuencias inmutables y ordenadas de objetos.
tupla = (1, "a", 3.0, True)
```

### Tipos de Mapeo

```python
# dict: Colección de pares clave-valor. Mutable y, desde Python 3.7, mantiene el orden de inserción.
diccionario = {"nombre": "Maxi", "edad": 30}
```

### Conjuntos (Sets)

```python
# set: Colección no ordenada de elementos únicos. Mutable.
conjunto = {1, 2, 3, 3, 4}

# frozenset: Versión inmutable de un set.
conjunto_inmutable = frozenset([1, 2, 3])
```

### Booleanos y Nulos

```python
# bool: Representa valores de verdad (True o False).
es_valido = True

# NoneType: Tipo del objeto None, que representa la ausencia de valor.
sin_valor = None
```

## Parte 2: Referencia Exhaustiva de Palabras Reservadas

### `and`
Operador lógico AND.

### `as`
Se usa para crear un alias al importar módulos (`import math as m`) o en el manejo de excepciones (`except Exception as e`).

### `assert`
Se utiliza para depuración. Comprueba una condición y lanza `AssertionError` si es falsa.

### `async`
Declara una función como una corrutina (para programación asíncrona).

### `await`
Pausa la ejecución de una corrutina hasta que un objeto `awaitable` se complete.

### `break`
Sale del bucle más interno.

### `class`
Define una clase.

### `continue`
Salta a la siguiente iteración de un bucle.

### `def`
Define una función o un método.

### `del`
Elimina un objeto o una referencia.

### `elif`
Parte de una estructura `if` para una condición alternativa ('else if').

### `else`
Parte de una estructura `if` o `try` que se ejecuta si la condición principal no se cumple o no hay excepciones.

### `except`
Bloque que captura y maneja excepciones dentro de un `try`.

### `False`
Literal booleano para falso.

### `finally`
Bloque que se ejecuta siempre al salir de un `try`, haya o no excepción.

### `for`
Bucle para iterar sobre una secuencia.

### `from`
Se usa con `import` para importar partes de un módulo.

### `global`
Declara que una variable dentro de una función se refiere a una variable del ámbito global.

### `if`
Inicia una estructura de control condicional.

### `import`
Importa un módulo.

### `in`
Operador de pertenencia. Comprueba si un valor está en una secuencia.

### `is`
Operador de identidad. Comprueba si dos variables apuntan al mismo objeto en memoria.

### `lambda`
Crea una función anónima de una sola línea.

### `nonlocal`
Declara que una variable se refiere a una variable en el ámbito de una función anidada superior (no global).

### `None`
Representa la ausencia de valor.

### `not`
Operador de negación lógica.

### `or`
Operador lógico OR.

### `pass`
Operación nula. Se usa como marcador de posición.

### `raise`
Lanza una excepción.

### `return`
Sale de una función y devuelve un valor.

### `True`
Literal booleano para verdadero.

### `try`
Inicia un bloque de manejo de excepciones.

### `while`
Inicia un bucle que se ejecuta mientras una condición sea verdadera.

### `with`
Envuelve la ejecución de un bloque con métodos definidos por un context manager.

### `yield`
Pausa una función generadora y devuelve un valor.
