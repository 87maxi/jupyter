# Programación Orientada a Objetos (OOP) Avanzada en Python

Python es un lenguaje profundamente orientado a objetos. Este notebook explora conceptos avanzados como la herencia múltiple, el uso de `super()`, las propiedades y los métodos mágicos (dunder methods) que permiten personalizar el comportamiento de los objetos.

## 1. Clases, Atributos y Métodos

En Python, los atributos de clase son compartidos por todas las instancias, mientras que los atributos de instancia son únicos para cada objeto.

```python
class Usuario:
    # Atributo de Clase
    plataforma = "Web"

    def __init__(self, nombre, rol):
        # Atributos de Instancia
        self.nombre = nombre
        self.rol = rol
        self.__id_interno = 1234  # Atributo "Privado" (Mangling de nombres)

    # Método de Instancia
    def describir(self):
        return f"Usuario {self.nombre} con rol {self.rol} en {self.plataforma}"

    # Método de Clase (Accede a la clase, no a la instancia)
    @classmethod
    def cambiar_plataforma(cls, nueva):
        cls.plataforma = nueva

    # Método Estático (No accede a la clase ni a la instancia)
    @staticmethod
    def es_valido(nombre):
        return len(nombre) > 3

u = Usuario("Maxi", "Admin")
print(u.describir())
print(f"ID Interno: {u._Usuario__id_interno}") # Acceso al atributo 'privado' mediante mangling
```

## 2. Herencia y Polimorfismo

Python soporta herencia simple y múltiple. El orden de resolución de métodos se rige por el **MRO (Method Resolution Order)**.

```python
class Empleado(Usuario):
    def __init__(self, nombre, rol, salario):
        # Llama al constructor de la superclase
        super().__init__(nombre, rol)
        self.salario = salario

    # Sobrescritura de método (Override)
    def describir(self):
        desc_base = super().describir()
        return f"{desc_base} y gana {self.salario}"

e = Empleado("Ana", "Dev", 5000)
print(e.describir())
```

## 3. Propiedades (@property)

Permiten definir métodos que se comportan como atributos, facilitando la validación y el encapsulamiento.

```python
class Circulo:
    def __init__(self, radio):
        self._radio = radio

    @property
    def radio(self):
        return self._radio

    @radio.setter
    def radio(self, valor):
        if valor < 0:
            raise ValueError("El radio no puede ser negativo")
        self._radio = valor

    @property
    def area(self):
        import math
        return math.pi * (self._radio ** 2)

c = Circulo(5)
print(f"Área: {c.area:.2f}")
c.radio = 10
print(f"Nueva Área: {c.area:.2f}")
```

## 4. Métodos Mágicos (Dunder Methods)

Son métodos que empiezan y terminan con doble guion bajo (`__`). Permiten que tus objetos se comporten como tipos integrados.

```python
class Libro:
    def __init__(self, titulo, autor, paginas):
        self.titulo = titulo
        self.autor = autor
        self.paginas = paginas

    # Representación para humanos
    def __str__(self):
        return f"'{self.titulo}' por {self.autor}"

    # Representación para desarrolladores (recreación del objeto)
    def __repr__(self):
        return f"Libro('{self.titulo}', '{self.autor}', {self.paginas})"

    # Longitud del objeto
    def __len__(self):
        return self.paginas

    # Sobrecarga de operador suma (+)
    def __add__(self, otro):
        return self.paginas + otro.paginas

l1 = Libro("El Quijote", "Cervantes", 1000)
l2 = Libro("Python Pro", "Guido", 300)

print(l1)
print(f"Páginas totales: {l1 + l2}")
print(f"Longitud l1: {len(l1)}")
```

## 5. Clases Abstractas

Se usan para definir una interfaz que las subclases deben implementar obligatoriamente.

```python
from abc import ABC, abstractmethod

class Forma(ABC):
    @abstractmethod
    def calcular_area(self):
        pass

class Cuadrado(Forma):
    def __init__(self, lado):
        self.lado = lado
    
    def calcular_area(self):
        return self.lado * self.lado

# f = Forma() # Error: Can't instantiate abstract class Forma
cuad = Cuadrado(4)
print(f"Área Cuadrado: {cuad.calcular_area()}")
```

## 6. Mixins y Herencia Múltiple

Los Mixins son clases diseñadas para proporcionar funcionalidad adicional a otras clases mediante herencia múltiple, sin ser instanciadas por sí mismas.

```python
class LoggerMixin:
    def log(self, mensaje):
        print(f"[LOG - {self.__class__.__name__}]: {mensaje}")

class BaseDeDatos(LoggerMixin):
    def conectar(self):
        self.log("Conectando a la DB...")

db = BaseDeDatos()
db.conectar()
```
