# Iteradores, Generadores y Decoradores en Python

Este notebook explora tres de los conceptos más potentes y elegantes de Python. Estas herramientas permiten escribir código más eficiente, legible y modular, facilitando el manejo de grandes volúmenes de datos y la reutilización de lógica.

## 1. Iteradores

Un iterador es un objeto que permite recorrer una colección de elementos uno a uno. En Python, un objeto es un iterador si implementa los métodos mágicos `__iter__()` y `__next__()`.

```python
class Contador:
    def __init__(self, limite):
        self.limite = limite
        self.actual = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.actual < self.limite:
            valor = self.actual
            self.actual += 1
            return valor
        else:
            raise StopIteration

for n in Contador(3):
    print(n)
```

## 2. Generadores

Los generadores son una forma sencilla de crear iteradores usando la palabra clave `yield`. A diferencia de las funciones normales, un generador mantiene su estado entre ejecuciones, permitiendo la **evaluación perezosa** (lazy evaluation), lo que ahorra mucha memoria.

```python
def generador_fibonacci(limite):
    a, b = 0, 1
    count = 0
    while count < limite:
        yield a
        a, b = b, a + b
        count += 1

fib = generador_fibonacci(5)
for num in fib:
    print(num)

# Expresiones generadoras (Sintaxis similar a list comprehension)
cuadrados_gen = (x**2 for x in range(1000000)) # No ocupa memoria para el millón de elementos
```

## 3. Decoradores

Un decorador es una función que envuelve a otra función para extender su funcionalidad sin modificar su código fuente.

```python
import functools

def depurar(func):
    @functools.wraps(func) # Mantiene los metadatos de la función original
    def wrapper(*args, **kwargs):
        print(f"Llamando a {func.__name__} con argumentos {args} {kwargs}")
        resultado = func(*args, **kwargs)
        print(f"{func.__name__} retornó {resultado}")
        return resultado
    return wrapper

@depurar
def sumar(a, b):
    return a + b

sumar(5, 10)
```

## 4. Decoradores con Argumentos

Para pasar argumentos a un decorador, necesitamos crear una función adicional que actúe como factoría de decoradores.

```python
def repetir(veces):
    def decorador_real(func):
        @functools.wraps(func)
        def wrapper(*args, **kwargs):
            for _ in range(veces):
                resultado = func(*args, **kwargs)
            return resultado
        return wrapper
    return decorador_real

@repetir(veces=3)
def saludar(nombre):
    print(f"Hola {nombre}")

saludar("Mundo")
```

## 5. Decoradores de Clase

También se pueden aplicar decoradores a clases enteras o definir decoradores como clases usando el método `__call__`.

```python
class ContadorLlamadas:
    def __init__(self, func):
        self.func = func
        self.llamadas = 0

    def __call__(self, *args, **kwargs):
        self.llamadas += 1
        print(f"Llamada número {self.llamadas} a {self.func.__name__}")
        return self.func(*args, **kwargs)

@ContadorLlamadas
def test():
    print("Hola")

test()
test()
```

## 6. Resumen de Casos de Uso

| Herramienta | Cuándo usar |
| :--- | :--- |
| **Iterador** | Cuando necesitas definir una forma personalizada de recorrer una clase propia. |
| **Generador** | Para procesar flujos de datos grandes o infinitos sin cargar todo en memoria. |
| **Decorador** | Para añadir lógica transversal como logging, autenticación, caché o medición de tiempo. |
