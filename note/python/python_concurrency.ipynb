# Concurrencia y Paralelismo en Python

Python ofrece tres enfoques principales para manejar la ejecución simultánea. La elección depende de si la tarea está limitada por la CPU o por operaciones de entrada/salida (I/O).

## 1. Threading (Hilos)

Los hilos son ideales para tareas **I/O bound** (esperar red, archivos o bases de datos). Debido al **GIL (Global Interpreter Lock)**, Python solo ejecuta un hilo de código Python a la vez, por lo que no sirven para paralelismo real en cálculos intensivos de CPU.

```python
import threading
import time

def tarea_hilo(nombre):
    print(f"Hilo {nombre}: iniciando")
    time.sleep(2)
    print(f"Hilo {nombre}: finalizado")

hilo1 = threading.Thread(target=tarea_hilo, args=("A",))
hilo1.start()
hilo1.join() # Esperar a que termine
```

## 2. Multiprocessing (Multiprocesamiento)

El módulo `multiprocessing` crea procesos independientes, cada uno con su propio intérprete de Python y espacio de memoria. Esto permite **paralelismo real** en sistemas multinúcleo, ideal para tareas **CPU bound** (procesamiento de imágenes, cálculos matemáticos).

```python
from multiprocessing import Process

def calculo_pesado(n):
    return sum(i * i for i in range(n))

if __name__ == "__main__":
    p = Process(target=calculo_pesado, args=(1000000,))
    p.start()
    p.join()
```

## 3. Asyncio (Programación Asíncrona)

Usa un **Event Loop** (bucle de eventos) para manejar miles de conexiones concurrentes en un solo hilo. Es extremadamente eficiente para servidores web y sockets. Usa las palabras clave `async` y `await`.

```python
import asyncio

async def tarea_asincrona(id):
    print(f"Tarea {id} iniciando")
    await asyncio.sleep(1) # No bloquea el hilo
    print(f"Tarea {id} terminada")

async def main():
    # Ejecutar varias tareas concurrentemente
    await asyncio.gather(
        tarea_asincrona(1),
        tarea_asincrona(2)
    )

# En un notebook, se suele ejecutar directamente:
# await main()
```

## 4. El GIL (Global Interpreter Lock)

El GIL es un mutex que protege el acceso a los objetos de Python, evitando que múltiples hilos ejecuten bytecodes de Python al mismo tiempo. 

- **Impacto**: Los hilos en Python son concurrentes pero no paralelos para tareas de computación.
- **Solución**: Usar `multiprocessing` o librerías en C (como NumPy) que liberan el GIL durante cálculos pesados.

## 5. Resumen de Elección

| Modelo | Mejor para... | Ventaja | Desventaja |
| :--- | :--- | :--- | :--- |
| **Threading** | I/O (Red, Disco) | Ligero, comparte memoria | Bloqueado por el GIL |
| **Multiprocessing** | CPU (Cálculos) | Paralelismo real | Consumo de memoria alto |
| **Asyncio** | Red (Miles de conexiones) | Muy escalable en un solo hilo | Curva de aprendizaje, código viral |
