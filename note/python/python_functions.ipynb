# Funciones y Programación Funcional en Python

Las funciones son los bloques de construcción fundamentales en Python. Permiten encapsular lógica, reutilizar código y estructurar aplicaciones de forma modular. Python también soporta paradigmas de programación funcional de primer nivel.

## 1. Definición y Argumentos

Las funciones se definen con la palabra clave `def`. Python ofrece una gran flexibilidad en cómo se pasan los argumentos.

```python
def saludar(nombre, mensaje="Hola"): # Parámetro con valor por defecto
    return f"{mensaje}, {nombre}!"

print(saludar("Maxi"))
print(saludar("Ana", "Bienvenida")) # Sobrescribiendo el default

# Argumentos Variables (*args y **kwargs)
def sumar_todo(*numeros): 
    # *numeros es una tupla de todos los argumentos posicionales extra
    return sum(numeros)

def configurar(**opciones):
    # **opciones es un diccionario de todos los argumentos de clave extra
    for clave, valor in opciones.items():
        print(f"{clave}: {valor}")

print(sumar_todo(1, 2, 3, 4, 5))
configurar(puerto=8080, debug=True, timeout=30)
```

## 2. Retorno de Valores

En Python, si no se especifica un `return`, la función devuelve `None`. Se pueden devolver múltiples valores usando tuplas de forma implícita.

```python
def calcular_min_max(lista):
    return min(lista), max(lista) # Retorna una tupla

minimo, maximo = calcular_min_max([10, 2, 55, 1]) # Unpacking
print(f"Mínimo: {minimo}, Máximo: {maximo}")
```

## 3. Ámbito de Variables (Scope)

Python sigue la regla **LEGB** para la resolución de nombres:
1. **Local**: Dentro de la función.
2. **Enclosing**: En funciones anidadas.
3. **Global**: Nivel de módulo.
4. **Built-in**: Funciones integradas del lenguaje.

```python
global_var = "Soy global"

def funcion_externa():
    enclosing_var = "Soy de la función padre"
    
    def funcion_interna():
        local_var = "Soy local"
        print(local_var)
        print(enclosing_var) # Acceso a scope superior
        print(global_var)
    
    funcion_interna()

funcion_externa()
```

## 4. Programación Funcional

### Lambda Functions
Funciones anónimas de una sola línea.

```python
duplicar = lambda x: x * 2
print(duplicar(5))

# Uso común en ordenamiento
puntos = [(1, 2), (4, 1), (5, -3)]
puntos.sort(key=lambda p: p[1]) # Ordenar por el segundo elemento
print(puntos)
```

### Funciones de Orden Superior: Map, Filter y Reduce

```python
from functools import reduce

numeros = [1, 2, 3, 4, 5, 6]

# map: Aplicar función a cada elemento
dobles = list(map(lambda x: x * 2, numeros))

# filter: Filtrar elementos según condición
pares = list(filter(lambda x: x % 2 == 0, numeros))

# reduce: Acumular valores
suma_total = reduce(lambda x, y: x + y, numeros)

print(f"Dobles: {dobles}")
print(f"Pares: {pares}")
print(f"Suma Total: {suma_total}")
```

## 5. Decoradores

Un decorador es una función que recibe otra función y extiende su comportamiento sin modificarla explícitamente.

```python
def temporizador(func):
    import time
    def wrapper(*args, **kwargs):
        inicio = time.time()
        resultado = func(*args, **kwargs)
        fin = time.time()
        print(f"Tiempo de ejecución de {func.__name__}: {fin - inicio:.4f} seg")
        return resultado
    return wrapper

@temporizador
def operacion_lenta():
    import time
    time.sleep(1)
    print("Operación finalizada")

operacion_lenta()
```

## 6. Type Hinting (Anotaciones de Tipo)

Aunque Python es dinámico, las anotaciones de tipo ayudan a la documentación y a herramientas de análisis estático como `mypy`.

```python
def procesar_usuario(nombre: str, edad: int) -> str:
    return f"Usuario {nombre} de {edad} años procesado."

from typing import List, Dict, Optional

def obtener_promedio(notas: List[float]) -> float:
    return sum(notas) / len(notas)

def buscar_id(db: Dict[int, str], id: int) -> Optional[str]:
    return db.get(id)
```
