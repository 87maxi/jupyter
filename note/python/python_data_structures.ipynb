# Estructuras de Datos y Colecciones en Python

Python ofrece un conjunto rico de estructuras de datos integradas que son extremadamente versátiles. Además de los tipos básicos, el módulo `collections` proporciona alternativas especializadas para casos de uso avanzados.

## 1. Listas (Lists)

Las listas son colecciones ordenadas y mutables de elementos. Pueden contener tipos de datos heterogéneos.

```python
lista = [1, "Python", 3.14, True]

# Métodos comunes
lista.append("nuevo")    # Añadir al final
lista.insert(1, "medio") # Insertar en índice
lista.pop()             # Eliminar y retornar último
lista.remove("Python")  # Eliminar por valor

# Slicing (Rebanado)
sub_lista = lista[1:3]   # Elementos desde índice 1 hasta 2
reversa = lista[::-1]    # Invertir lista
```

## 2. Tuplas (Tuples)

Similares a las listas pero **inmutables**. Se usan para datos que no deben cambiar o como claves de diccionarios (ya que son hashables).

```python
punto = (10, 20)
# punto[0] = 15 # Error: 'tuple' object does not support item assignment

# Unpacking (Desempaquetado)
x, y = punto

# Tupla de un solo elemento (requiere coma)
single = (5,)
```

## 3. Diccionarios (Dictionaries)

Almacenan pares clave-valor. A partir de Python 3.7+, mantienen el orden de inserción.

```python
usuario = {"id": 1, "nombre": "Maxi", "rol": "Admin"}

# Acceso seguro
rol = usuario.get("rol", "Invitado") # Retorna default si no existe

# Iteración
for k, v in usuario.items():
    print(f"{k}: {v}")

# Unir diccionarios (Python 3.9+)
perfil = {"bio": "Ingeniero"}
completo = usuario | perfil
```

## 4. Conjuntos (Sets)

Colecciones no ordenadas de elementos únicos. Útiles para eliminar duplicados y realizar operaciones matemáticas de conjuntos.

```python
a = {1, 2, 3, 4}
b = {3, 4, 5, 6}

print(a | b) # Unión: {1, 2, 3, 4, 5, 6}
print(a & b) # Intersección: {3, 4}
print(a - b) # Diferencia: {1, 2}
print(a ^ b) # Diferencia simétrica: {1, 2, 5, 6}
```

## 5. El Módulo `collections` (Avanzado)

Proporciona tipos de datos alternativos.

```python
from collections import namedtuple, deque, Counter, defaultdict

# 1. namedtuple: Tuplas con nombres de campo
Punto = namedtuple('Punto', ['x', 'y'])
p = Punto(10, 20)
print(p.x, p.y)

# 2. deque: Cola doblemente terminada (eficiente para pop/append en ambos extremos)
cola = deque(["a", "b"])
cola.appendleft("z")
cola.pop()

# 3. Counter: Cuenta ocurrencias de elementos
letras = Counter("mississippi")
print(letras.most_common(2))

# 4. defaultdict: Diccionario que inicializa valores si no existen
conteo = defaultdict(int)
conteo["clicks"] += 1 # No lanza KeyError
```

## 6. Slicing y Operaciones de Secuencia

Python permite manipular secuencias de forma muy expresiva.

```python
nums = list(range(10))

# [inicio:fin:paso]
pares = nums[::2]
ultimos_tres = nums[-3:]
copia = nums[:] # Crea una copia superficial (shallow copy)
```
