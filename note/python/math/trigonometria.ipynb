```python

import plotly.express as px
import pandas as pd


import math

## polinomio ###
po = lambda i: math.sin( math.radians(i) ) 
## end polinomio ###

r = 1000


df = pd.DataFrame(dict(
    x = [ i for i in range(-r, r)  ],
    y = [ po(i) for i in range(-r, r)  ]
))
fig = px.line(df, x="x", y="y", title="Seno") 
fig.show()
```
