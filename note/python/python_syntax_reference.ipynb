# Referencia Completa de Sintaxis de Python

Python es un lenguaje de programación de alto nivel, interpretado y de propósito general. Su filosofía de diseño enfatiza la legibilidad del código con el uso de una sangría significativa.

## 1. Variables y Tipos de Datos

Python es dinámicamente tipado. No es necesario declarar el tipo de una variable al crearla.

```python
# Enteros y Flotantes
entero = 10
flotante = 10.5

# Strings (Cadenas de texto)
nombre = "Python"
multilinea = """Esta es una cadena
de varias líneas"""

# Booleanos
es_cierto = True
es_falso = False

# None (Nulo)
vacio = None

# Interpolación de strings (f-strings, Python 3.6+)
print(f"Hola, {nombre}. El valor es {flotante}")
```

## 2. Estructuras de Datos Integradas

- **Listas**: Ordenadas, mutables.
- **Tuplas**: Ordenadas, inmutables.
- **Diccionarios**: Pares clave-valor, mutables.
- **Sets**: Colección no ordenada de elementos únicos.

```python
# Listas
frutas = ["manzana", "banana", "cereza"]
frutas.append("naranja")
primera = frutas[0]

# Tuplas
punto = (10, 20)
x, y = punto # Desempaquetado

# Diccionarios
usuario = {"nombre": "Maxi", "edad": 30}
print(usuario["nombre"])

# Sets
unicos = {1, 2, 2, 3} # {1, 2, 3}
```

## 3. Control de Flujo

```python
# if / elif / else
edad = 18
if edad >= 18:
    print("Adulto")
elif edad > 12:
    print("Adolescente")
else:
    print("Niño")

# Bucle for (Iterar sobre colecciones)
for fruta in frutas:
    print(fruta)

# Range
for i in range(5): # 0 a 4
    print(i)

# Bucle while
contador = 0
while contador < 3:
    print(f"Contador: {contador}")
    contador += 1
```

## 4. Comprensiones (Comprehensions)

Forma concisa de crear listas, diccionarios y sets.

```python
# List Comprehension
cuadrados = [x**2 for x in range(10) if x % 2 == 0]

# Dictionary Comprehension
mapa_cuadrados = {x: x**2 for x in range(5)}
```

## 5. Funciones

Se definen con `def`. Soportan argumentos por defecto, variables (`*args`) y clave-valor (`**kwargs`).

```python
def saludar(nombre, mensaje="Hola"):
    return f"{mensaje}, {nombre}"

def funcion_flexible(*args, **kwargs):
    print(args)   # Tupla de argumentos posicionales
    print(kwargs) # Diccionario de argumentos de clave

# Funciones Lambda (Anónimas)
sumar = lambda a, b: a + b
print(sumar(5, 3))
```

## 6. Programación Orientada a Objetos (OOP)

```python
class Animal:
    def __init__(self, nombre):
        self.nombre = nombre # Atributo de instancia

    def hablar(self):
        pass

class Perro(Animal): # Herencia
    def hablar(self):
        return f"{self.nombre} dice ¡Guau!"

mi_perro = Perro("Fido")
print(mi_perro.hablar())
```

## 7. Manejo de Errores

```python
try:
    resultado = 10 / 0
except ZeroDivisionError as e:
    print(f"Error: {e}")
except Exception as e:
    print(f"Error inesperado: {e}")
finally:
    print("Finalizando operación")
```

## 8. Gestión de Archivos y Context Managers

```python
# La sentencia 'with' asegura que el archivo se cierre automáticamente
with open("ejemplo.txt", "w") as f:
    f.write("Hola Python")

with open("ejemplo.txt", "r") as f:
    contenido = f.read()
    print(contenido)
```

## 9. Decoradores y Generadores

Conceptos avanzados para modificar comportamiento y manejar flujos de datos eficientes.

```python
# Decorador simple
def log_operacion(func):
    def wrapper(*args, **kwargs):
        print("Iniciando ejecución...")
        return func(*args, **kwargs)
    return wrapper

@log_operacion
def mi_funcion():
    print("Hola desde la función")

# Generador (Lazy evaluation)
def contador_infinito():
    n = 0
    while True:
        yield n
        n += 1
```
