# Tipos y Definiciones en Rust: Una Guía Exhaustiva

Rust ofrece una variedad de palabras clave para definir tipos y estructuras que permiten modelar datos y comportamientos de forma precisa. Este notebook explora cada una de ellas, desde las estructuras básicas hasta las abstracciones avanzadas.

## 1. Definición de Datos: `struct`, `enum` y `union`

### A. `struct` (Estructuras)
Se usan para agrupar datos relacionados. Existen tres variantes:
1. **Clásica**: Con campos nombrados.
2. **Tupla**: Campos identificados por posición.
3. **Unit-like**: Sin campos (útiles para implementar traits en tipos que no necesitan datos).

```rust
// Estructura clásica
struct Usuario {
    id: u64,
    nombre: String,
}

// Estructura de tupla (Newtype pattern común)
struct Color(i32, i32, i32);

// Estructura unitaria
struct Marcador;
```

### B. `enum` (Enumeraciones)
Permiten definir un tipo que puede ser una de varias variantes. En Rust, las variantes pueden contener datos, lo que los convierte en tipos de datos algebraicos (ADT).

```rust
enum Evento {
    Click(i32, i32),          // Variante con tupla
    Tecla { codigo: String }, // Variante con campos nombrados
    Cerrar,                   // Variante simple
}
```

### C. `union` (Uniones)
Similares a las de C. Se usan principalmente para interoperabilidad con C (FFI). Son inherentemente **unsafe** porque el compilador no sabe qué variante es válida en cada momento.

```rust
union MiUnion {
    f1: u32,
    f2: f32,
}

let u = MiUnion { f1: 1 };
unsafe {
    println!("f1: {}", u.f1);
}
```

## 2. Comportamiento y Abstracción: `trait` e `impl`

### A. `trait` (Rasgos)
Definen una interfaz de comportamiento que otros tipos pueden implementar. Pueden tener métodos con implementaciones por defecto.

```rust
trait Volador {
    fn despegar(&self);
    
    // Implementación por defecto
    fn aterrizar(&self) {
        println!("Aterrizando...");
    }
}
```

### B. `impl` (Implementaciones)
Se usa para añadir métodos a una estructura o enumeración, o para implementar un trait en un tipo.

```rust
struct Avion;

// Métodos propios
impl Avion {
    fn nuevo() -> Self { Avion }
}

// Implementar un trait
impl Volador for Avion {
    fn despegar(&self) {
        println!("Avión despegando");
    }
}
```

## 3. Alias y Genéricos: `type` y `<T>`

### A. `type` (Alias de Tipo)
Crea un nombre alternativo para un tipo existente. No crea un tipo nuevo (a diferencia de una struct), solo un alias.

```rust
type Kilometros = i32;
type ResultadoOperacion<T> = Result<T, String>;

let distancia: Kilometros = 100;
```

### B. Genéricos (`<T>`)
Permiten definir funciones o tipos que funcionan con múltiples tipos de datos, reduciendo la duplicación.

```rust
struct Contenedor<T> {
    valor: T,
}

fn procesar<T: std::fmt::Display>(item: T) {
    println!("{}", item);
}
```

## 4. Tipos Propios de Rust (Avanzado)

### A. `impl Trait` vs `dyn Trait` (Polimorfismo)

1. **`impl Trait` (Static Dispatch)**: El compilador resuelve el tipo exacto en tiempo de compilación. Es muy eficiente.
2. **`dyn Trait` (Dynamic Dispatch)**: El tipo se resuelve en tiempo de ejecución usando *vtable*. Requiere estar detrás de un puntero (`&dyn` o `Box<dyn>`).

```rust
// Retorna un tipo que implementa Iterator, pero no decimos cuál exactamente
fn obtener_iterador() -> impl Iterator<Item = i32> {
    vec![1, 2, 3].into_iter()
}

// Recibe una referencia a CUALQUIER tipo que implemente Volador
fn volar_objeto(v: &dyn Volador) {
    v.despegar();
}
```

### B. `Self` (Mayúscula) vs `self` (Minúscula)
- `self`: El objeto instancia actual (parámetro de método).
- `Self`: El tipo que está implementando el bloque `impl` o el `trait`.

```rust
impl Avion {
    // Self es equivalente a Avion aquí
    fn copiar(&self) -> Self {
        Avion
    }
}
```

## 5. Resumen de Cuándo Usar Qué

| Definición | Cuándo usarla |
| :--- | :--- |
| **`struct`** | Cuando tienes un conjunto de datos fijos que "pertenecen" juntos. |
| **`enum`** | Cuando un dato puede ser una cosa **U** otra (Excluyente). |
| **`trait`** | Cuando quieres definir qué puede hacer un tipo, permitiendo abstracción. |
| **`type`** | Para simplificar firmas de tipos largos o dar significado semántico. |
| **`mod`** | Para organizar el código en archivos y carpetas. |
| **`impl`** | Para dar funcionalidad a tus tipos. |
| **`union`** | Solo para FFI avanzado con C/C++. |
| **`impl Trait`** | Para retornar tipos complejos u ocultar detalles de implementación. |
| **`dyn Trait`** | Cuando necesitas una colección de diferentes tipos que comparten un trait. |
