# Palabras Reservadas y Tipos de Datos en Rust: Una Referencia Exhaustiva

Este notebook sirve como una guía completa de la sintaxis fundamental de Rust. Se divide en dos secciones principales:
1.  **Tipos de Datos Fundamentales**: Explica los tipos primitivos y compuestos sobre los que se construye el lenguaje.

## Parte 1: Tipos de Datos Fundamentales

Rust es un lenguaje de **tipado estático**, lo que significa que el compilador debe conocer el tipo de cada variable en tiempo de compilación. Esto previene una categoría entera de errores antes de que el programa se ejecute.

### Tipos Escalares
Representan un único valor.

```rust
// --- Enteros (Integers) ---
// Con signo (i) y sin signo (u). El número indica los bits.
let con_signo: i32 = -10;       // i8, i16, i32 (default), i64, i128
let sin_signo: u32 = 10;        // u8, u16, u32, u64, u128
let de_arquitectura: isize = 0; // Depende de la arquitectura (32 o 64 bits)

// --- Punto Flotante (Floating-Point) ---
let flotante_32: f32 = 3.14;
let flotante_64: f64 = 6.28; // Default para flotantes

// --- Booleanos (Booleans) ---
let es_rust: bool = true;

// --- Caracteres (Characters) ---
// Un solo carácter Unicode (4 bytes), se usan comillas simples.
let letra: char = 'a';
let emoji: char = '🦀';
```

### Tipos Compuestos
Agrupan múltiples valores.

```rust
// --- Tuplas (Tuples) ---
// Longitud fija, pueden contener diferentes tipos.
let tupla: (i32, f64, char) = (100, 3.14, 'A');

// Desestructuración
let (x, y, z) = tupla;
println!("El valor de y es: {}", y);

// Acceso por índice
let primer_valor = tupla.0;

// --- Arreglos (Arrays) ---
// Longitud fija, todos los elementos deben ser del mismo tipo.
// Se almacenan en el Stack.
let array_numeros: [i32; 5] = [1, 2, 3, 4, 5];
let primer_numero = array_numeros[0];

// Inicializar un array con el mismo valor
let array_ceros = [0; 10]; // Diez ceros
```

## Parte 2: Referencia Exhaustiva de Palabras Reservadas

### `as`
1.  Realizar una conversión de tipo (casting) insegura.
2.  Renombrar un ítem importado con `use`.

```rust
let a = 10u8;
let b = a as f64; // Casting

use std::collections::HashMap as MiMapa; // Alias
```

### `async`
Marca una función como asíncrona, haciendo que retorne un `Future`.

```rust
async fn tarea_lenta() -> u32 {
    // ... operaciones asíncronas ...
    42
}
```

### `await`
Se usa dentro de una función `async` para pausar la ejecución hasta que un `Future` se complete.

```rust
async fn ejecutar() {
    let resultado = tarea_lenta().await;
    println!("Resultado: {}", resultado);
}
```

### `break`
Sale inmediatamente del bucle más interno. Puede retornar un valor en un `loop`.

```rust
let mut contador = 0;
let resultado = loop {
    contador += 1;
    if contador == 5 {
        break contador * 2; // Retorna 10
    }
};
```

### `const`
Declara una constante. Requiere una anotación de tipo explícita y su valor debe ser conocido en tiempo de compilación.

```rust
const PI: f64 = 3.14159;
```

### `continue`
Salta a la siguiente iteración de un bucle.

```rust
for i in 1..5 {
    if i % 2 == 0 { continue; } // Saltar números pares
    println!("{}", i);
}
```

### `crate`
Referencia a la raíz del crate actual. Útil para rutas absolutas dentro del crate.

```rust
// mod mi_modulo {
//     pub fn mi_funcion() {}
// }
// use crate::mi_modulo::mi_funcion;
```

### `dyn`
Indica que se está utilizando un *trait object* para polimorfismo dinámico.

```rust
trait Imprimible {
    fn imprimir(&self);
}
let vector: Vec<Box<dyn Imprimible>> = vec![];
```

### `else`
Parte de una estructura `if`/`else` que se ejecuta si la condición `if` es falsa.

```rust
if 1 > 2 {
    // ...
} else {
    println!("1 no es mayor que 2");
}
```

### `enum`
Define un tipo enumerado, que puede ser una de varias variantes.

```rust
enum Color {
    Rojo,
    Verde,
    Azul,
}
```

### `extern`
Se utiliza para interactuar con código escrito en otros lenguajes (Foreign Function Interface - FFI).

```rust
// extern "C" {
//     fn abs(input: i32) -> i32;
// }
```

### `false`
Literal booleano que representa falsedad.

### `fn`
Declara una nueva función.

```rust
fn saludar() {
    println!("Hola!");
}
```

### `for`
Itera sobre los elementos de una colección o rango.

```rust
for i in 1..=3 { // Rango inclusivo
    println!("Número: {}", i);
}
```

### `if`
Ejecuta código condicionalmente.

### `impl`
Implementa funcionalidades (métodos o traits) para un tipo.

```rust
struct Circulo { radio: f64 }
impl Circulo {
    fn area(&self) -> f64 {
        3.14 * self.radio * self.radio
    }
}
```

### `in`
Se utiliza en bucles `for` para iterar sobre un `Iterator`.

### `let`
Declara una variable local. Por defecto, inmutable.

### `loop`
Crea un bucle infinito.

### `match`
Compara un valor contra una serie de patrones de forma exhaustiva.

```rust
let valor = 2;
match valor {
    1 => println!("Uno"),
    2 => println!("Dos"),
    _ => println!("Otro"),
}
```

### `mod`
Declara un módulo para organizar el código.

### `move`
Fuerza que una clausura tome posesión de las variables capturadas.

```rust
let s = String::from("hola");
let closure = move || println!("{}", s);
closure();
// s ya no es válida aquí
```

### `mut`
Modificador para hacer una variable mutable.

### `pub`
Modificador para hacer un ítem público y accesible desde fuera de su módulo.

### `ref`
Se utiliza en *pattern matching* para tomar una referencia a un valor.

```rust
let opcion = Some(5);
if let Some(ref val) = opcion {
    println!("Referencia: {}", val);
}
```

### `return`
Sale de la función actual y devuelve un valor.

### `self` (minúscula)
Se refiere a la instancia actual de un `struct` o `enum` en un método.

### `Self` (mayúscula)
Se refiere al tipo que se está implementando en un bloque `impl` o `trait`.

### `static`
Declara una variable global con duración de vida `'static`.

```rust
static CONTADOR: i32 = 0;
```

### `struct`
Define una estructura personalizada para agrupar datos.

### `super`
Se refiere al módulo padre del módulo actual.

### `trait`
Define un conjunto de comportamientos que los tipos pueden implementar.

### `true`
Literal booleano que representa veracidad.

### `type`
Crea un alias para un tipo existente.

### `union`
Define un tipo de unión, similar a C. Es `unsafe`.

### `unsafe`
Marca un bloque de código donde se pueden realizar operaciones que el compilador no puede garantizar como seguras.

### `use`
Importa rutas al ámbito actual.

### `where`
Especifica límites de trait para genéricos de forma más clara.

```rust
use std::fmt::Debug;
fn mi_funcion<T>(t: T) where T: Debug {}
```

### `while`
Crea un bucle que se ejecuta mientras una condición sea verdadera.
