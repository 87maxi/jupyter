## Parte 1: Tipos de Datos Fundamentales

En Ruby, todo es un objeto. Cada valor, desde un nÃšmero hasta `nil`, es una instancia de una clase y tiene mÃĐtodos asociados.

### Tipos NumÃĐricos

```ruby
# Integer: NÃšmeros enteros de cualquier tamaÃąo.
entero = 100
numero_grande = 1_000_000_000_000

# Float: NÃšmeros de punto flotante.
flotante = 3.14

# Rational: NÃšmeros racionales (fracciones).
racional = Rational(3, 4) # 3/4

# Complex: NÃšmeros complejos.
complejo = Complex(1, 2) # 1 + 2i
```

### Texto y SÃ­mbolos

```ruby
# String: Cadenas de texto mutables.
cadena_interpolada = "Hola, #{100}" # Comillas dobles permiten interpolaciÃģn
cadena_literal = 'Hola, #{100}'      # Comillas simples no interpolan

# Symbol: Identificadores inmutables y eficientes. Muy usados como claves de Hashes.
simbolo = :mi_clave
puts simbolo.object_id == :mi_clave.object_id # => true
```

### Booleanos y `nil`

```ruby
verdadero = true  # Instancia de TrueClass
falso = false     # Instancia de FalseClass
nulo = nil        # Instancia de NilClass, representa la ausencia de valor
```

### Colecciones

```ruby
# Array: ColecciÃģn ordenada y mutable de objetos.
mi_array = [1, "dos", 3.0, :cuatro]
mi_array << 5 # AÃąade un elemento

# Hash: ColecciÃģn de pares clave-valor (diccionario).
mi_hash = { nombre: "Maxi", edad: 30 } # Sintaxis moderna
otro_hash = { "clave" => "valor" }    # Sintaxis con 'hashrocket'

# Range: Representa un intervalo de valores.
rango_inclusivo = (1..5)  # 1, 2, 3, 4, 5
rango_exclusivo = (1...5) # 1, 2, 3, 4
```

### `alias`
Crea un alias (un nombre alternativo) para un mÃĐtodo o variable global existente.

```ruby
class Saludo
  def hola; "Hola!"; end
  alias :hi :hola
end
puts Saludo.new.hi
```

### `and`
Operador lÃģgico AND con baja precedencia. Se prefiere `&&` para expresiones booleanas complejas.

### `begin`

### `BEGIN`
Define un bloque que se ejecuta una sola vez, antes de que cualquier otro cÃģdigo del script se ejecute.

```ruby
BEGIN { puts "Esto se ejecuta primero." }
```

### `break`

### `case` / `when` / `else` / `then`
Estructura de control para mÃšltiples condiciones, similar a un `switch`.

```ruby
valor = 2
case valor
when 1 then puts "Uno"
when 2 then puts "Dos"
else puts "Otro"
end
```
