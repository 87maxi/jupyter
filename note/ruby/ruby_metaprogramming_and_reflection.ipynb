# Metaprogramación y Reflexión en Ruby

Ruby es uno de los lenguajes más potentes en cuanto a metaprogramación (escribir código que escribe código). Gran parte de la magia de frameworks como Ruby on Rails se basa en estas técnicas.

## 1. Introspección (Reflexión)

Ruby permite examinar el estado de los objetos y las clases en tiempo de ejecución.

```ruby
class Ejemplo
  def hola; "Hola"; end
end

obj = Ejemplo.new

puts "Clase: #{obj.class}"
puts "¿Es instancia de Ejemplo?: #{obj.is_a?(Ejemplo)}"
puts "Métodos definidos: #{obj.methods.grep(/ho/)}"
puts "¿Responde a :hola?: #{obj.respond_to?(:hola)}"
```

## 2. Envío Dinámico: `send`

El método `send` permite llamar a un método por su nombre (como símbolo o string), incluso si es privado.

```ruby
class Calculadora
  def sumar(a, b); a + b; end
end

calc = Calculadora.new
metodo = :sumar
puts calc.send(metodo, 10, 5) # Equivale a calc.sumar(10, 5)
```

## 3. `method_missing`

Cuando Ruby no encuentra un método, llama a `method_missing`. Sobrescribirlo permite crear APIs fluidas y dinámicas.

```ruby
class Fantasma
  def method_missing(nombre, *args, &bloque)
    puts "Intentaste llamar a '#{nombre}' con argumentos: #{args}"
    puts "¡Pero este método no existe, así que lo atrapo yo!"
  end

  # Es buena práctica sobrescribir también respond_to_missing?
  def respond_to_missing?(nombre, incluir_privado = false)
    true
  end
end

f = Fantasma.new
f.cualquier_cosa(1, 2, 3)
```

## 4. Definición Dinámica de Métodos: `define_method`

Permite crear métodos en tiempo de ejecución dentro de una clase.

```ruby
class CreadorDeMetodos
  [:lunes, :martes, :miercoles].each do |dia|
    define_method("hoy_es_#{dia}?") do
      puts "¿Hoy es #{dia}? Tal vez..."
    end
  end
end

creador = CreadorDeMetodos.new
creador.hoy_es_lunes?
```

## 5. `instance_eval` y `class_eval`

- **`instance_eval`**: Evalúa código en el contexto de una instancia (permite acceder a variables privadas).
- **`class_eval`**: Evalúa código en el contexto de una clase (permite añadir métodos a la clase desde fuera).

```ruby
class Secreto
  def initialize; @dato = "Contraseña123"; end
end

s = Secreto.new
# Rompiendo la encapsulación con instance_eval
puts s.instance_eval { @dato }

Secreto.class_eval do
  def nuevo_metodo
    "Añadido dinámicamente"
  end
end

puts s.nuevo_metodo
```

## 6. Ganchos (Hooks)

Ruby avisa a tu código cuando ocurren eventos en el ciclo de vida de los objetos.

```ruby
module Rastreador
  def self.included(base)
    puts "#{base} incluyó el módulo Rastreador"
  end
end

class MiClase
  include Rastreador
end

# Otros hooks comunes: inherited (para herencia), method_added, extended.
```
