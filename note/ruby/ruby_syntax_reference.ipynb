# Referencia Completa de Sintaxis de Ruby

Ruby es un lenguaje de programación dinámico, reflexivo y orientado a objetos, enfocado en la simplicidad y la productividad. Este notebook cubre desde los fundamentos hasta las características más avanzadas del lenguaje.

## 1. Variables y Tipos de Datos

En Ruby, todo es un objeto. Las variables no necesitan declaración de tipo (tipado dinámico).

```ruby
# 1. Variables Locales (empiezan con minúscula o _)
nombre = "Ruby"
edad = 25

# 2. Variables de Instancia (empiezan con @)
@precio = 100.50

# 3. Variables de Clase (empiezan con @@)
@@contador = 0

# 4. Variables Globales (empiezan con $)
$version = "3.2"

# 5. Constantes (empiezan con MAYÚSCULA)
PI = 3.14159

puts "Nombre: #{nombre}, Edad: #{edad}" # Interpolación de strings
```

## 2. Strings y Símbolos

- **Strings**: Mutables, encerrados en `""` o `''`.
- **Símbolos**: Inmutables, identificadores únicos. Se usan comúnmente como claves de hash.

```ruby
str = "Hola"
sym = :hola

puts str.object_id == "Hola".object_id # false (diferentes objetos)
puts sym.object_id == :hola.object_id # true (mismo objeto)

# Métodos útiles
puts "ruby".capitalize # "Ruby"
puts "  espacios  ".strip # "espacios"
```

## 3. Colecciones: Arrays y Hashes

```ruby
# Arrays
numeros = [1, 2, 3, 4]
numeros << 5 # Append
numeros.push(6)

# Hashes (Diccionarios)
# Sintaxis moderna (JSON-like)
usuario = { nombre: "Maxi", edad: 30 }

# Sintaxis clásica (hashrocket)
opciones = { "color" => "azul", :tamaño => "grande" }

puts usuario[:nombre] # Acceso por símbolo
puts opciones["color"] # Acceso por string
```

## 4. Estructuras de Control

Ruby ofrece una sintaxis muy natural para el control de flujo.

```ruby
# if / else
lluvia = true
if lluvia
  puts "Lleva paraguas"
elsif false
  puts "No llueve"
else
  puts "Día soleado"
end

# Modificador if (sintaxis corta)
puts "Es verdadero" if true

# unless (equivalente a if !)
puts "No ha terminado" unless false

# Case (Switch)
puntuacion = 85
case puntuacion
when 90..100 then puts "Sobresaliente"
when 70..89  then puts "Notable"
else puts "A mejorar"
end
```

## 5. Bucles e Iteradores

En Ruby, preferimos iteradores sobre bucles `for` tradicionales.

```ruby
# Each
[1, 2, 3].each do |num|
  puts "Numero: #{num}"
end

# Times
3.times { puts "Repetición" }

# Map (Transformación)
cuadrados = [1, 2, 3].map { |n| n * n }

# Select (Filtrado)
pares = [1, 2, 3, 4, 5].select { |n| n.even? }
```

## 6. Métodos

La última expresión evaluada es el valor de retorno implícito.

```ruby
def saludar(nombre = "Invitado")
  "Hola, #{nombre}!"
end

puts saludar("Maxi")

# Parámetros nombrados (Keyword arguments)
def configurar(host:, puerto: 80)
  puts "Conectando a #{host}:#{puerto}"
end

configurar(host: "localhost", puerto: 3000)
```

## 7. Clases y Objetos

Ruby es estrictamente orientado a objetos.

```ruby
class Animal
  # Genera getters y setters automáticamente
  attr_accessor :nombre

  def initialize(nombre)
    @nombre = nombre
  end

  def hablar
    "Haciendo sonidos..."
  end
end

class Perro < Animal # Herencia
  def hablar
    "Guau!"
  end
end

mi_perro = Perro.new("Fido")
puts mi_perro.nombre
puts mi_perro.hablar
```

## 8. Módulos y Mixins

Los módulos sirven para agrupar métodos relacionados o como "Mixins" para añadir funcionalidad a clases sin usar herencia.

```ruby
module Volador
  def volar
    "Estoy volando!"
  end
end

class Pajaro < Animal
  include Volador # Mixin
end

p = Pajaro.new("Piolín")
puts p.volar
```

## 9. Bloques, Procs y Lambdas

Ruby permite pasar bloques de código como argumentos.

```ruby
# Yield (ejecuta el bloque pasado)
def ejecutar_bloque
  puts "Antes"
  yield if block_given?
  puts "Después"
end

ejecutar_bloque { puts "Durante" }

# Lambdas
mi_lambda = ->(x) { x * 2 }
puts mi_lambda.call(5)
```

## 10. Manejo de Errores

Usa `begin`, `rescue`, `ensure`.

```ruby
begin
  # Operación peligrosa
  1 / 0
rescue ZeroDivisionError => e
  puts "Error detectado: #{e.message}"
ensure
  puts "Esta parte siempre se ejecuta"
end
```
