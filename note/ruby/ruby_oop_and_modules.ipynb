# Programación Orientada a Objetos (OOP) y Módulos en Ruby

Ruby es un lenguaje de programación puramente orientado a objetos. Todo valor en Ruby es un objeto, incluyendo los números y los valores booleanos. Este notebook profundiza en cómo estructurar aplicaciones utilizando clases, herencia y módulos.

## 1. Clases y Atributos

Las clases se definen con la palabra clave `class`. Los atributos se manejan mediante variables de instancia (`@`) y métodos de acceso.

```ruby
class Usuario
  # Accesores automáticos
  attr_accessor :nombre  # Lectura y escritura
  attr_reader :id        # Solo lectura

  def initialize(id, nombre)
    @id = id
    @nombre = nombre
  end

  # Método de instancia
  def presentarse
    "Hola, soy #{@nombre} con ID #{@id}"
  end
end

u = Usuario.new(1, "Maxi")
puts u.presentarse
u.nombre = "Maximiliano"
puts u.nombre
```

## 2. Herencia y `super`

Ruby permite herencia simple. Una subclase hereda todos los métodos de su superclase.

```ruby
class Empleado < Usuario
  attr_accessor :puesto

  def initialize(id, nombre, puesto)
    super(id, nombre) # Llama al constructor de la superclase
    @puesto = puesto
  end

  def presentarse
    super + " y trabajo como #{@puesto}"
  end
end

emp = Empleado.new(2, "Ana", "Desarrolladora")
puts emp.presentarse
```

## 3. Visibilidad de Métodos

Ruby tiene tres niveles de visibilidad:
- **Public**: Accesible desde cualquier lugar (por defecto).
- **Protected**: Accesible solo por la clase y sus subclases.
- **Private**: Accesible solo internamente dentro del objeto.

```ruby
class CajaFuerte
  def initialize(codigo)
    @codigo = codigo
  end

  def abrir
    validar_codigo
    "Caja abierta"
  end

  private

  def validar_codigo
    puts "Validando..."
  end
end

caja = CajaFuerte.new("1234")
puts caja.abrir
# caja.validar_codigo # Esto causaría un NoMethodError (es privado)
```

## 4. Métodos y Variables de Clase

Los métodos de clase actúan sobre la clase misma, no sobre instancias.

```ruby
class Contador
  @@total = 0 # Variable de clase

  def self.incrementar
    @@total += 1
  end

  def self.total
    @@total
  end

  def initialize
    Contador.incrementar
  end
end

Contador.new
Contador.new
puts "Total instancias: #{Contador.total}"
```

## 5. Módulos y Mixins

Los módulos tienen dos funciones principales:
1. **Espacios de nombres (Namespacing)**: Evitar colisiones de nombres.
2. **Mixins**: Simular herencia múltiple inyectando funcionalidad.

```ruby
module Autenticable
  def login(user, pass)
    puts "Autenticando a #{user}..."
  end
end

class Administrador < Usuario
  include Autenticable # Añade métodos como métodos de instancia
end

class SistemaExterno
  extend Autenticable # Añade métodos como métodos de clase
end

admin = Administrador.new(10, "SuperAdmin")
admin.login("admin", "secret") # Método de instancia

SistemaExterno.login("api", "key") # Método de clase
```

## 6. Módulos como Namespaces

Se usa el operador `::` para acceder a elementos dentro de un módulo.

```ruby
module Procesamiento
  class Imagen
    def self.info; "Clase Imagen de Procesamiento"; end
  end
end

module Redes
  class Imagen
    def self.info; "Clase Imagen de Redes"; end
  end
end

puts Procesamiento::Imagen.info
puts Redes::Imagen.info
```

## 7. Singleton Classes (Metaclases)

Ruby permite añadir métodos a una instancia específica sin afectar a otras instancias de la misma clase.

```ruby
str1 = "Hola"
str2 = "Mundo"

def str1.gritar
  self.upcase + "!!!"
end

puts str1.gritar
# puts str2.gritar # Error: str2 no tiene el método gritar
```
