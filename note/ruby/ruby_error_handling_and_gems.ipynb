# Manejo de Errores y Ecosistema de Gemas en Ruby

Ruby utiliza un sistema de excepciones para el manejo de errores. Además, cuenta con un potente sistema de gestión de librerías llamado RubyGems y herramientas de empaquetado como Bundler.

## 1. Excepciones: `begin`, `rescue`, `ensure` y `else`

El flujo básico para capturar errores es envolver el código propenso a fallos en un bloque `begin`.

````ruby
def dividir(a, b)
  begin
    resultado = a / b
  rescue ZeroDivisionError => e
    puts "Error: No puedes dividir por cero (#{e.message})"
    nil
  rescue StandardError => e
    puts "Error inesperado: #{e.class}"
    nil
  else
    puts "La operación fue exitosa"
    resultado
  ensure
    puts "Cerrando recursos..."
  end
end

puts "Resultado: #{dividir(10, 0)}"
````

## 2. Lanzar Excepciones: `raise` y `fail`

Puedes lanzar tus propias excepciones usando `raise`. Por convención, `raise` se usa para excepciones y `fail` es un alias semántico.

````ruby
class EdadInvalidaError < StandardError; end

def validar_edad(edad)
  raise EdadInvalidaError, "La edad debe ser mayor a 0" if edad <= 0
  puts "Edad válida: #{edad}"
end

begin
  validar_edad(-5)
rescue EdadInvalidaError => e
  puts "Capturado: #{e.message}"
end
````

## 3. Reintentar: `retry`

Ruby permite reintentar un bloque de código desde el principio si ocurre una excepción. **Cuidado con los bucles infinitos.**

````ruby
intentos = 0
begin
  intentos += 1
  puts "Intento ##{intentos}..."
  raise "Fallo de red" if intentos < 3
  puts "Éxito en el intento final!"
rescue => e
  retry if intentos < 5
  puts "Se rindió después de demasiados intentos"
end
````

## 4. RubyGems y Bundler

- **Gema**: Un paquete de Ruby (librería o ejecutable).
- **RubyGems**: El gestor de paquetes de Ruby.
- **Bundler**: Herramienta para gestionar dependencias de un proyecto (basado en un archivo `Gemfile`).

### El archivo `Gemfile` (Ejemplo)

```ruby
source 'https://rubygems.org'

gem 'rails', '~> 7.0'
gem 'pg', '>= 1.1'
gem 'nokogiri'

group :development, :test do
  gem 'rspec'
  gem 'pry'
end
```

### Comandos comunes:
- `gem install <nombre>`: Instala una gema globalmente.
- `bundle install`: Instala las gemas definidas en el `Gemfile`.
- `bundle exec <comando>`: Ejecuta un comando en el contexto de las gemas del proyecto.

## 5. Carga de Código: `require` vs `load`

- **`require`**: Carga una librería solo una vez (detecta duplicados). Se usa para librerías instaladas.
- **`require_relative`**: Carga un archivo relativo al archivo actual.
- **`load`**: Vuelve a cargar el archivo cada vez que se llama (útil en desarrollo para ver cambios sin reiniciar).
