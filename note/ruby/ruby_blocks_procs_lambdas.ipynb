# Bloques, Procs y Lambdas en Ruby

Ruby es famoso por su manejo de la programación funcional mediante cierres (closures). En Ruby, esto se manifiesta de tres formas principales: Bloques, Procs y Lambdas. Entender la diferencia entre ellos es clave para escribir código idiomático.

## 1. Bloques

Los bloques son trozos de código que se pasan a un método. No son objetos, pero pueden ser ejecutados por el método usando la palabra clave `yield`.

```ruby
def mi_iterador
  puts "Inicio del método"
  yield("Valor 1") if block_given?
  yield("Valor 2") if block_given?
  puts "Fin del método"
end

mi_iterador do |val|
  puts "Recibido en el bloque: #{val}"
end
```

## 2. Procs (Procedimientos)

Un `Proc` es un objeto que encapsula un bloque de código, permitiendo guardarlo en una variable o pasarlo como argumento a otros métodos.

```ruby
saludo = Proc.new { |nombre| puts "Hola #{nombre}" }

saludo.call("Ruby")
saludo.call("Rails")

# Convertir bloque en Proc usando el símbolo &
def metodo_con_proc(&bloque)
  puts "Clase del bloque: #{bloque.class}"
  bloque.call
end

metodo_con_proc { puts "Ejecutando bloque convertido" }
```

## 3. Lambdas

Las lambdas son un tipo especial de `Proc`. Aunque son similares, tienen dos diferencias fundamentales con los Procs normales.

```ruby
# Sintaxis de flecha (moderna)
mi_lambda = ->(x, y) { x + y }

# Sintaxis clásica
otra_lambda = lambda { |x| x * 2 }

puts mi_lambda.call(10, 5)
puts otra_lambda.call(10)
```

## 4. Diferencias: Proc vs Lambda

### A. Manejo de Argumentos
- **Lambda**: Lanza un error si el número de argumentos es incorrecto (estricto).
- **Proc**: Ignora argumentos extra o asigna `nil` a los faltantes (permisivo).

```ruby
p = Proc.new { |x, y| puts "x: #{x}, y: #{y}" }
l = ->(x, y) { puts "x: #{x}, y: #{y}" }

p.call(1)       # Funciona (y es nil)
# l.call(1)     # Error: wrong number of arguments
```

### B. Comportamiento del `return`
- **Lambda**: El `return` sale solo de la lambda y vuelve al método llamador.
- **Proc**: El `return` intenta salir del método donde se definió el Proc (lo que puede causar errores si el ámbito original ya no existe).

```ruby
def test_return
  l = lambda { return "Retorno de Lambda" }
  res = l.call
  puts "Después de la lambda: #{res}"

  p = Proc.new { return "Retorno de Proc" }
  p.call
  puts "Esto nunca se verá porque el Proc salió del método"
end

puts test_return
```

## 5. El patrón Enumerable y Bloques

Ruby utiliza bloques extensivamente para iterar colecciones mediante el módulo `Enumerable`.

```ruby
numeros = [1, 2, 3, 4, 5]

# find
primero_grande = numeros.find { |n| n > 3 }

# inject (reduce)
suma = numeros.inject(0) { |acum, n| acum + n }

puts "Suma: #{suma}"
```
