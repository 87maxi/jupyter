# Módulos y Paquetes en Go

Go utiliza un sistema de paquetes y módulos para organizar el código y gestionar las dependencias. Este notebook explica cómo estructurar proyectos en Go y cómo funciona la visibilidad entre paquetes.

## 1. Módulos (`go mod`)

Un **Módulo** es una colección de paquetes relacionados que se versionan juntos. Se define mediante un archivo `go.mod` en la raíz del proyecto.

### Comandos básicos:
- `go mod init <nombre-modulo>`: Inicializa un nuevo módulo.
- `go mod tidy`: Limpia dependencias no usadas y añade las faltantes.
- `go mod download`: Descarga dependencias al caché local.

## 2. Paquetes (`package`)

Cada archivo de Go debe pertenecer a un paquete. Todos los archivos en el mismo directorio deben declarar el mismo nombre de paquete.

### Reglas de Oro:
1. El nombre del paquete suele coincidir con el nombre del directorio.
2. El paquete `main` es el punto de entrada para ejecutables.
3. Los nombres de los paquetes deben ser cortos, concisos y en minúsculas.

````go
// Archivo: calculadora/operaciones.go
package calculadora

// Sumar es exportada (Empieza con Mayúscula)
func Sumar(a, b int) int {
    return a + b
}

// restar es privada (Empieza con minúscula)
func restar(a, b int) int {
    return a - b
}
````

## 3. Visibilidad (Exported vs Unexported)

Go usa una regla extremadamente simple para la visibilidad:

- **Exportado**: Si el identificador (variable, función, struct, tipo) comienza con una **letra Mayúscula**, es visible desde otros paquetes.
- **No Exportado**: Si comienza con una **letra minúscula**, solo es visible dentro del mismo paquete.

````go
type Usuario struct {
    Nombre string // Campo exportado
    email  string // Campo privado (no visible fuera del paquete)
}

func (u Usuario) ObtenerEmail() string {
    return u.email // Ok: el método está en el mismo paquete
}
````

## 4. Importación de Paquetes

Para usar código de otro paquete, se utiliza la sentencia `import`.

````go
import (
    "fmt"                  // Paquete estándar
    "math/rand"            // Paquete estándar anidado
    "github.com/google/uuid" // Paquete externo
    "mi-proyecto/util"     // Paquete local (dentro del mismo módulo)
)

// Alias de importación
import f "fmt"

// Importación para efectos secundarios (solo ejecuta init())
import _ "github.com/lib/pq"
````

## 5. La función `init`

Cada paquete puede tener una o más funciones `init()`. Estas se ejecutan automáticamente cuando se carga el paquete, antes de la función `main` y después de que las variables globales del paquete hayan sido inicializadas.

````go
var dbConexion string

func init() {
    dbConexion = "Conectado a la DB"
    fmt.Println("Inicializando paquete...")
}
````

## 6. Paquetes Internos (`internal`)

Go tiene una convención especial: si un paquete está dentro de un directorio llamado `internal`, solo puede ser importado por paquetes que comparten el mismo ancestro común. Esto es útil para ocultar lógica que no quieres exponer como API pública del módulo.

## 7. Estructura de Proyecto Recomendada

```text
mi-proyecto/
├── go.mod
├── go.sum
├── main.go             // Punto de entrada
├── api/                // Lógica de red
│   └── router.go
├── internal/           // Código privado del proyecto
│   └── auth/
│       └── logic.go
└── pkg/                // Código que otros proyectos podrían usar (opcional)
    └── util/
        └── strings.go
```
