## Parte 1: Tipos de Datos Fundamentales

Go es un lenguaje de **tipado estático**. El tipo de una variable se conoce en tiempo de compilación, lo que ayuda a prevenir errores.

### Tipos Básicos
Representan valores individuales.

```go
package main
import "fmt"

func main() {
    // --- Booleanos ---
    var esValido bool = true

    // --- Numéricos ---
    // Enteros con signo (int8, int16, int32, int64, int)
    var entero int = -100
    // Enteros sin signo (uint8, uint16, uint32, uint64, uint)
    var sinSigno uint = 100
    // Punto flotante (float32, float64)
    var flotante float64 = 3.14159
    // Complejos (complex64, complex128)
    var complejo complex128 = 2 + 3i

    // --- Strings ---
    // Cadenas de texto inmutables, codificadas en UTF-8.
    var saludo string = "Hola, Go!"

    // --- Runas y Bytes ---
    // byte es un alias para uint8.
    var miByte byte = 'a'
    // rune es un alias para int32 y representa un punto de código Unicode.
    var miRuna rune = '⌘'

    fmt.Println(esValido, entero, flotante, complejo, saludo, miByte, miRuna)
}
```

### Tipos Compuestos
Agrupan múltiples valores.

```go
package main
import "fmt"

func main() {
    // --- Arrays ---
    // Longitud fija, todos los elementos deben ser del mismo tipo.
    var miArray [3]int = [3]int{1, 2, 3}

    // --- Slices ---
    // Vista dinámica y flexible sobre un array subyacente. Son más comunes que los arrays.
    var miSlice []int = []int{10, 20, 30}
    miSlice = append(miSlice, 40)

    // --- Structs ---
    // Agrupan campos de diferentes tipos.
    type Persona struct {
        Nombre string
        Edad   int
    }
    var p Persona = Persona{Nombre: "Ana", Edad: 30}

    // --- Maps ---
    // Pares clave-valor no ordenados.
    var miMapa map[string]int = make(map[string]int)
    miMapa["clave"] = 100

    fmt.Println(miArray, miSlice, p, miMapa)
}
```

## Parte 2: Referencia Exhaustiva de Palabras Reservadas

### `break`
Sale del bucle `for` o `switch` más interno.

### `case`
Una rama dentro de un `switch` o `select`.

### `chan`
Declara un canal, que es un conducto tipado para comunicarse entre goroutines.

### `const`
Declara una constante.

### `continue`
Salta a la siguiente iteración del bucle `for`.

### `default`
Cláusula por defecto en un `switch` o `select`.

### `defer`
Pospone la ejecución de una función hasta que la función circundante retorne.

### `else`
Parte de una estructura `if`/`else` que se ejecuta si la condición es falsa.

### `fallthrough`
En un `switch`, transfiere el control a la siguiente cláusula `case`.

### `for`
El único constructor de bucles en Go.

### `func`
Declara una función.

### `go`
Lanza una nueva goroutine.

### `goto`
Transfiere el control a una etiqueta. Su uso es desaconsejado.

### `if`
Ejecuta código condicionalmente.

### `import`
Importa un paquete.

### `interface`
Define un conjunto de firmas de métodos.

### `map`
Define un tipo de mapa (clave-valor).

### `package`
Declara el paquete al que pertenece el archivo.

### `range`
Se utiliza con `for` para iterar sobre slices, arrays, strings, maps o channels.

### `return`
Sale de una función y, opcionalmente, devuelve valores.

### `select`
Permite esperar en múltiples operaciones de canal.

### `struct`
Define una estructura, un tipo compuesto.

### `switch`
Estructura de control para múltiples condiciones.

### `type`
Define un nuevo tipo.

### `var`
Declara una o más variables.
