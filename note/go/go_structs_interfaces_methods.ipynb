# Estructuras, Métodos e Interfaces en Go

Go no es un lenguaje orientado a objetos en el sentido tradicional (no tiene clases ni herencia de tipos). Sin embargo, permite definir estructuras de datos, asociarles comportamiento mediante métodos y definir polimorfismo mediante interfaces.

## 1. Structs (Estructuras)

Una `struct` es una colección de campos. Es la forma básica de agrupar datos.

```go
type Rectangulo struct {
    Ancho  float64
    Altura float64
}

// Instanciación
r1 := Rectangulo{Ancho: 10, Altura: 5}
r2 := Rectangulo{10, 5} // Orden posicional
r3 := Rectangulo{}      // Valores por defecto (0.0)

// Structs Anónimas (útiles para pruebas o datos temporales)
config := struct {
    APIKey string
    Puerto int
}{
    APIKey: "abc-123",
    Puerto: 8080,
}
```

## 2. Métodos y Receptores

Un método es una función con un argumento especial llamado **receptor**. El receptor aparece entre la palabra clave `func` y el nombre del método.

```go
// Receptor de Valor (Copia el objeto)
func (r Rectangulo) Area() float64 {
    return r.Ancho * r.Altura
}

// Receptor de Puntero (Puede modificar el objeto original y evita copias)
func (r *Rectangulo) Escalar(factor float64) {
    r.Ancho *= factor
    r.Altura *= factor
}

// REGLA: Usa receptores de puntero si:
// 1. Necesitas modificar el estado.
// 2. La estructura es muy grande.
// 3. Quieres consistencia (usar siempre punteros).
```

## 3. Composición mediante Embebido (Embedding)

Go favorece la composición sobre la herencia. Puedes "embeber" un tipo en otro para heredar sus campos y métodos de forma automática.

```go
type Persona struct {
    Nombre string
}

func (p Persona) Saludar() {
    fmt.Printf("Hola, soy %s\n", p.Nombre)
}

type Desarrollador struct {
    Persona // Campo embebido (sin nombre)
    Lenguaje string
}

dev := Desarrollador{
    Persona: Persona{Nombre: "Maxi"},
    Lenguaje: "Go",
}

dev.Saludar() // Se accede directamente al método embebido
```

## 4. Interfaces

Una interfaz es un conjunto de firmas de métodos. Un tipo implementa una interfaz implementando sus métodos. **No hay una declaración explícita `implements`**.

```go
type Dibujable interface {
    Dibujar() string
}

func (r Rectangulo) Dibujar() string {
    return "Dibujando un rectángulo"
}

// Función que acepta cualquier cosa que implemente la interfaz
func Renderizar(d Dibujable) {
    fmt.Println(d.Dibujar())
}

// Interfaz Vacía: interface{} o any (Go 1.18+)
// Puede contener cualquier valor.
func Describir(i any) {
    fmt.Printf("(%v, %T)\n", i, i)
}
```

## 5. Type Assertions y Type Switches

Cómo recuperar el tipo original desde una interfaz.

```go
var i any = "hola"

// Assertion (Pánico si falla)
s := i.(string)

// Assertion Segura
s, ok := i.(string)
if !ok {
    fmt.Println("No es un string")
}

// Type Switch
switch v := i.(type) {
case int:
    fmt.Println("Es un entero", v)
case string:
    fmt.Println("Es un string", v)
default:
    fmt.Println("Tipo desconocido")
}
```
