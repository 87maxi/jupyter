# Referencia Completa de Sintaxis de Go (Golang)

Este notebook sirve como una guía detallada de la sintaxis y características del lenguaje de programación Go. Go es conocido por su simplicidad, eficiencia y excelente soporte para la concurrencia.

## 1. Variables y Constantes

Go ofrece varias formas de definir variables, con un enfoque en la legibilidad y el tipado estático.

```go
package main

import "fmt"

func main() {
    // 1. Declaración con 'var' (con tipo explícito)
    var x int = 10

    // 2. Inferencia de tipo
    var y = "Hola"

    // 3. Declaración corta (solo dentro de funciones)
    z := 20.5

    // 4. Múltiples variables
    var a, b int = 1, 2
    c, d := true, "mundo"

    // 5. Constantes (no pueden usar :=)
    const Pi = 3.14159

    fmt.Println(x, y, z, a, b, c, d, Pi)
}
```

## 3. Colecciones: Arrays, Slices y Maps

Go diferencia claramente entre arreglos de tamaño fijo y rodajas (slices) dinámicas.

```go
// Arrays (Tamaño fijo)
var arr [5]int = [5]int{1, 2, 3, 4, 5}

// Slices (Dinámicos - los más usados)
slice := []int{10, 20, 30}
slice = append(slice, 40) // Añadir elementos

// make([]tipo, longitud, capacidad)
s := make([]string, 3)

// Maps (Diccionarios)
m := make(map[string]int)
m["clave"] = 100
valor, existe := m["clave"] // Acceso seguro
```

## 4. Control de Flujo

Go solo tiene una estructura de bucle: `for`. No existe `while` ni `do-while`.

```go
// If con sentencia corta
if v := 10; v > 5 {
    fmt.Println("Mayor")
} else {
    fmt.Println("Menor")
}

// For clásico
for i := 0; i < 5; i++ {
    fmt.Println(i)
}

// For como While
n := 1
for n < 5 {
    n++
}

// Range (para iterar colecciones)
nums := []int{1, 2, 3}
for indice, valor := range nums {
    fmt.Printf("Indice: %d, Valor: %d\n", indice, valor)
}

// Switch (no necesita break)
switch os := "linux"; os {
case "linux":
    fmt.Println("Pingüino")
default:
    fmt.Println("Otro")
}
```

## 5. Funciones

Go permite retornos múltiples y funciones como ciudadanos de primera clase.

```go
// Función con múltiples retornos
func dividir(a, b float64) (float64, error) {
    if b == 0 {
        return 0, fmt.Errorf("división por cero")
    }
    return a / b, nil
}

// Retornos nombrados
func rect(base, altura float64) (area float64) {
    area = base * altura
    return // Retorna 'area' automáticamente
}

// Closures (Funciones anónimas)
sumador := func(x int) int {
    return x + x
}
```

## 6. Estructuras (Structs) y Métodos

Go no tiene clases, pero usa estructuras y receptores para simular el comportamiento de objetos.

```go
type Persona struct {
    Nombre string
    Edad   int
}

// Método con receptor de valor
func (p Persona) Saludar() {
    fmt.Printf("Hola, soy %s\n", p.Nombre)
}

// Método con receptor de puntero (para modificar el objeto)
func (p *Persona) CumplirAnios() {
    p.Edad++
}

// Composición (no herencia)
type Empleado struct {
    Persona // Campo embebido
    Sueldo  float64
}

e := Empleado{Persona: Persona{"Ana", 25}, Sueldo: 3000}
e.Saludar() // Acceso directo a métodos de Persona
```

## 7. Interfaces

En Go, las interfaces se implementan de forma **implícita**. Un tipo satisface una interfaz simplemente implementando sus métodos.

```go
type Forma interface {
    Area() float64
}

type Circulo struct { Radio float64 }

func (c Circulo) Area() float64 { 
    return 3.14 * c.Radio * c.Radio 
}

// Interfaz vacía (acepta cualquier tipo)
var i interface{} = "Cualquier cosa"

// Type Assertion
s, ok := i.(string)
```

## 8. Manejo de Errores

Go no usa excepciones (`try/catch`). En su lugar, los errores son valores que se retornan.

```go
// Patrón estándar de manejo de errores
resultado, err := dividir(10, 0)
if err != nil {
    fmt.Println("Error:", err)
    return
}

// Defer: Asegura que algo se ejecute al final de la función
// Muy usado para cerrar recursos
/*
file, _ := os.Open("test.txt")
defer file.Close()
*/
```

## 9. Punteros y Memoria

Go tiene punteros pero no aritmética de punteros (por seguridad).

```go
val := 42
ptr := &val // Obtener dirección
fmt.Println(*ptr) // Desreferenciar

// new vs make
p := new(int) // reserva memoria y retorna *int (puntero)
slice := make([]int, 5) // inicializa tipos complejos (slice, map, chan)
```

## 10. Concurrencia (Goroutines y Channels)

La joya de Go. Es ligera y potente.

```go
// Iniciar hilo ligero (Goroutine)
go func() {
    fmt.Println("Asíncrono")
}()

// Canales (Canales de comunicación entre goroutines)
ch := make(chan int)

go func() {
    ch <- 10 // Enviar dato
}()

valor := <-ch // Recibir dato (bloqueante)

// Select (como switch pero para canales)
/*
select {
case msg1 := <-ch1:
    fmt.Println(msg1)
case ch2 <- valor:
    fmt.Println("Enviado")
}
*/
```

## 11. Genéricos (Go 1.18+)

Permiten escribir código que funciona con varios tipos manteniendo la seguridad de tipos.

```go
// Función genérica [T tipo]
func Sumar[T int | float64](a, b T) T {
    return a + b
}

// Uso
s1 := Sumar(5, 10)
s2 := Sumar(3.5, 2.1)
```
