# Módulos y Funciones en Elixir

En Elixir, los módulos son el medio para organizar funciones en espacios de nombres (namespaces). A diferencia de los lenguajes orientados a objetos, no hay clases; todo se basa en la transformación de datos a través de funciones organizadas en módulos.

## 1. Definición de Módulos

Los módulos se definen usando la macro `defmodule`. Por convención, los nombres de los módulos usan `PascalCase`.

```elixir
defmodule Calculadora do
  @moduledoc """
  Proporciona funciones para realizar operaciones matemáticas básicas.
  """

  @doc "Suma dos números."
  def sumar(a, b) do
    a + b
  end
end

Calculadora.sumar(5, 3)
```

## 2. Funciones Públicas y Privadas

- `def`: Define una función pública (accesible desde fuera del módulo).
- `defp`: Define una función privada (solo accesible dentro del mismo módulo).

```elixir
defmodule Saludador do
  def hola(nombre) do
    formatear_saludo(nombre)
  end

  defp formatear_saludo(nombre) do
    "Hola, #{nombre}!"
  end
end

Saludador.hola("Maxi")
# Saludador.formatear_saludo("Maxi") # Esto lanzaría un error
```

## 3. Cláusulas de Función y Guardas

Una característica potente de Elixir es que una función puede tener múltiples cuerpos basados en el *pattern matching* de sus argumentos. Las guardas (`when`) añaden condiciones adicionales.

```elixir
defmodule Factorial do
  # Caso base
  def calcular(0), do: 1
  
  # Caso recursivo con guarda
  def calcular(n) when n > 0 do
    n * calcular(n - 1)
  end
end

Factorial.calcular(5)
```

## 4. Argumentos por Defecto

Se utiliza la sintaxis `\\` para definir valores por defecto.

```elixir
defmodule Conexion do
  def conectar(host, puerto \\ 80) do
    "Conectando a #{host}:#{puerto}"
  end
end

IO.puts Conexion.conectar("localhost")
IO.puts Conexion.conectar("localhost", 443)
```

## 5. Directivas de Módulo

- `import`: Trae funciones de otro módulo al ámbito actual para llamarlas sin el prefijo.
- `alias`: Permite referirse a un módulo con un nombre más corto.
- `use`: Invoca una macro que inyecta código en el módulo actual (común en Phoenix o Ecto).

```elixir
defmodule MiApp.Utilidades.Cadenas do
  def gritar(texto), do: String.upcase(texto)
end

defmodule Ejemplo do
  alias MiApp.Utilidades.Cadenas, as: Cadenas
  import List, only: [first: 1]

  def probar do
    IO.puts Cadenas.gritar("hola")
    IO.puts first([1, 2, 3])
  end
end

Ejemplo.probar()
```

## 6. Funciones Anónimas

Se definen con `fn` y se invocan usando un punto `.` antes de los paréntesis.

```elixir
sumar = fn a, b -> a + b end
IO.puts sumar.(10, 5)

# Sintaxis corta (Capture operator &)
duplicar = &(&1 * 2)
IO.puts duplicar.(5)
```
