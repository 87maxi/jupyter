# Referencia Completa de Sintaxis de Elixir

Elixir es un lenguaje funcional, dinámico, diseñado para construir aplicaciones escalables y mantenibles. Se ejecuta sobre la máquina virtual de Erlang (BEAM), conocida por su baja latencia y tolerancia a fallos.

## 1. Tipos de Datos Básicos

En Elixir, los datos son inmutables. Una vez creado un valor, este no puede ser modificado.

```elixir
# Enteros y Flotantes
entero = 42
flotante = 3.14

# Átomos (Constantes cuyo nombre es su valor)
:ok
:error
:mi_atomo

# Booleanos (Son en realidad átomos :true y :false)
true == :true

# Strings (UTF-8, entre comillas dobles)
nombre = "Elixir"
interpolacion = "Hola #{nombre}"

# Charlists (Listas de enteros, entre comillas simples)
'abc' == [97, 98, 99]
```

## 2. Colecciones

### Listas vs Tuplas
- **Listas (`[]`)**: Almacenadas como listas enlazadas (acceso lineal). Ideales para procesamiento recursivo.
- **Tuplas (`{}`)**: Almacenadas de forma contigua en memoria (acceso rápido por índice). No recomendadas para colecciones grandes que cambian frecuentemente.

```elixir
# Listas
lista = [1, 2, 3]
lista = [0 | lista] # Concatenación por la cabeza (eficiente)
[cabeza | cola] = lista

# Tuplas
tupla = {:ok, "resultado"}
elem(tupla, 1) # "resultado"

# Maps (Diccionarios)
mapa = %{"clave" => "valor", :id => 1}
mapa[:id] # Acceso
mapa = %{mapa | id: 2} # Actualización (crea nuevo mapa)
```

## 3. Pattern Matching (El operador `=`)

En Elixir, `=` no es un operador de asignación, sino de coincidencia (match). Intenta igualar el lado izquierdo con el derecho.

```elixir
x = 1
1 = x # Match exitoso
# 2 = x # Error: MatchError

{:ok, mensaje} = {:ok, "Hola Mundo"}
IO.puts mensaje # "Hola Mundo"

# El operador pin (^)
# Se usa para comparar contra el valor actual de una variable en lugar de re-asignar
^x = 1
```

## 4. Control de Flujo

Elixir utiliza `case`, `cond` e `if`.

```elixir
# case: Pattern matching puro
case {:ok, "datos"} do
  {:ok, val} -> "Éxito: #{val}"
  {:error, _} -> "Error"
  _ -> "Cualquier otra cosa"
end

# cond: Similar a un if-else if
cond do
  2 + 2 == 5 -> "Falso"
  1 + 1 == 2 -> "Verdadero"
  true -> "Fallback"
end

# if / unless
if true do
  "Es verdad"
else
  "Es mentira"
end
```

## 5. El Operador Pipe (`|>`) 

Es la joya de Elixir. Pasa el resultado de la expresión anterior como primer argumento a la siguiente función.

```elixir
" ELIXIR "
|> String.trim()
|> String.downcase()
|> String.replace("e", "3")
# Resultado: "3lixir"
```

## 6. Módulos y Funciones

```elixir
defmodule Matematicas do
  @moduledoc "Módulo de ejemplo"

  # Función pública
  def sumar(a, b) do
    do_sumar(a, b)
  end

  # Función privada
  defp do_sumar(a, b), do: a + b

  # Funciones con múltiples cláusulas y guards
  def cero?(0), do: true
  def cero?(n) when is_integer(n), do: false
end

Matematicas.sumar(10, 5)
```

## 7. Enumerables y Listas por Comprensión

```elixir
Enum.map([1, 2, 3], fn x -> x * 2 end)
Enum.filter([1, 2, 3, 4], &(rem(&1, 2) == 0)) # Sintaxis corta &()

# Comprensiones
for n <- [1, 2, 3, 4], rem(n, 2) == 0, do: n * n
```

## 8. Procesos y Concurrencia (Modelo de Actores)

Los procesos en Elixir son extremadamente ligeros y no comparten memoria.

```elixir
parent = self()

# Crear un proceso (spawn)
pid = spawn(fn ->
  send(parent, {:hola, "desde el proceso hijo"})
end)

# Recibir mensajes
receive do
  {:hola, msg} -> IO.puts msg
after
  1_000 -> "Timeout"
end
```
