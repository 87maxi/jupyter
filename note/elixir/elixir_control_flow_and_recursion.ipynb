# Control de Flujo y Recursión en Elixir

Como lenguaje funcional, Elixir maneja el control de flujo de manera diferente a los lenguajes imperativos. No existen los bucles `for` o `while` tradicionales que mutan variables; en su lugar, se utiliza la **recursión** y el **pattern matching**.

## 1. Estructuras de Control

Aunque la recursión es preferida, Elixir proporciona estructuras para decisiones lógicas.

```elixir
# case: Compara un valor contra muchos patrones
resultado = {:ok, 200}

case resultado do
  {:ok, codigo} -> "Éxito con código #{codigo}"
  {:error, _} -> "Hubo un error"
  _ -> "Cualquier otro caso"
end
```

```elixir
# cond: Se usa cuando quieres verificar diferentes condiciones, no solo patrones
cond do
  2 + 2 == 5 -> "Esto no pasará"
  2 * 2 == 3 -> "Tampoco"
  1 + 1 == 2 -> "Esto es verdadero"
  true -> "Este es el else (siempre debe haber un true al final)"
end
```

```elixir
# if y unless
if true do
  "Es verdad"
else
  "No es verdad"
end

unless false do
  "Esto se ejecuta porque la condición es falsa"
end
```

## 2. Recursión: La forma de iterar en Elixir

Debido a la inmutabilidad de los datos, los bucles se realizan mediante funciones recursivas.

```elixir
defmodule Repetidor do
  def imprimir_n_veces(mensaje, n) when n <= 0 do
    :ok
  end

  def imprimir_n_veces(mensaje, n) do
    IO.puts(mensaje)
    imprimir_n_veces(mensaje, n - 1)
  end
end

Repetidor.imprimir_n_veces("Hola!", 3)
```

## 3. Mapas y Reducciones Recursivas

El patrón común es procesar una lista separando la cabeza (`head`) de la cola (`tail`).

```elixir
defmodule ListaUtil do
  # Sumar todos los elementos de una lista
  def sumar([]), do: 0
  def sumar([cabeza | cola]), do: cabeza + sumar(cola)

  # Duplicar cada elemento
  def duplicar([]), do: []
  def duplicar([cab | col]), do: [cab * 2 | duplicar(col)]
end

IO.puts ListaUtil.sumar([1, 2, 3, 4, 5])
IO.inspect ListaUtil.duplicar([1, 2, 3])
```

## 4. Tail Call Optimization (Optimización de llamada de cola)

Elixir optimiza las funciones recursivas si la llamada recursiva es lo **último** que hace la función. Esto evita que el stack crezca infinitamente.

```elixir
defmodule SumaOptimizada do
  def ejecutar(lista), do: do_sumar(lista, 0)

  # Al pasar el acumulador, la llamada recursiva es la última operación
  defp do_sumar([], acumulador), do: acumulador
  defp do_sumar([h | t], acc), do: do_sumar(t, acc + h)
end

SumaOptimizada.ejecutar([1, 10, 100])
```

## 5. Enum: El módulo estándar para iterar

En la práctica, rara vez escribirás recursión manual para colecciones; usarás el módulo `Enum`.

```elixir
lista = [1, 2, 3]

Enum.each(lista, fn x -> IO.puts(x) end)
nueva_lista = Enum.map(lista, fn x -> x * 10 end)
suma = Enum.reduce(lista, 0, fn x, acc -> x + acc end)

IO.inspect nueva_lista
IO.puts suma
```
