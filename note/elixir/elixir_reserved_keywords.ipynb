## Parte 1: Tipos de Datos Fundamentales

En Elixir, todos los datos son **inmutables**. Una vez que un valor es creado, no puede ser modificado. Las "actualizaciones" siempre crean una nueva copia del dato.

### Tipos Básicos

```elixir
# Integer: Números enteros de cualquier tamaño.
entero = 42

# Float: Números de punto flotante de 64 bits.
flotante = 3.14

# Atom: Constantes cuyo nombre es su propio valor. Son muy eficientes para comparaciones.
atomo = :ok
otro_atomo = :error_no_encontrado

# Boolean: Son simplemente los átomos :true y :false.
es_valido = true
```

### Cadenas de Texto

```elixir
# String: Secuencias de bytes codificadas en UTF-8. Se definen con comillas dobles.
cadena = "Hola, Elixir"
interpolada = "El valor es: #{10 * 2}"

# Charlist: Listas de puntos de código Unicode. Se definen con comillas simples. Menos comunes en el día a día.
lista_de_caracteres = 'hola' # Equivalente a [104, 111, 108, 97]
```

### Colecciones

```elixir
# List: Implementadas como listas enlazadas. La inserción en la cabeza es muy rápida.
lista = [1, 2, 3]
nueva_lista = [0 | lista] # [0, 1, 2, 3]

# Tuple: Almacenados en memoria contigua. El acceso por índice es muy rápido. Ideales para agrupar un número fijo de elementos.
tupla = {:ok, "Resultado exitoso"}

# Map: Colección de pares clave-valor. Es la estructura de datos más versátil para este propósito.
mapa = %{nombre: "Maxi", edad: 30}
mapa.nombre # "Maxi"

# Keyword List: Una lista especial de tuplas de dos elementos donde el primer elemento es un átomo. Permite claves duplicadas.
lista_de_claves = [opcion1: true, opcion2: false, opcion1: false]
```

## Parte 2: Referencia Exhaustiva de Palabras Reservadas

### `after`
Se usa en bloques `receive` para definir un tiempo de espera. Si no se recibe ningún mensaje que coincida en el tiempo especificado, se ejecuta el bloque `after`.

```elixir
receive do
  {:mensaje, _} -> IO.puts("Mensaje recibido!")
after
  1_000 -> IO.puts("No se recibió nada en 1 segundo.")
end
```

### `alias` (Macro)
Crea un alias para un módulo, permitiendo referirse a él con un nombre más corto.

### `and`
Operador lógico booleano estricto (ambos operandos deben ser booleanos). Devuelve un booleano.

### `case`
Evalúa una expresión y la compara con una serie de patrones. Ejecuta el bloque del primer patrón que coincide.

### `cond`
Evalúa múltiples condiciones booleanas, ejecutando el bloque de la primera que sea verdadera.

### `def` / `defp` (Macros)
Definen funciones. `def` para funciones públicas, `defp` para privadas.

### `defmacro` / `defmacrop` (Macros)
Definen macros. `defmacrop` para macros privadas.

### `defmodule` (Macro)
Define un nuevo módulo.

### `defprotocol` / `defimpl` (Macros)
Definen protocolos e implementaciones para polimorfismo.

### `do` / `end`
Delimitan un bloque de código.

### `else`
Cláusula por defecto en estructuras como `if` y `with`.

### `false`
Literal booleano para falso (el átomo `:false`).

### `fn`
Define una función anónima (lambda).

### `for`
Realiza una lista por comprensión (*list comprehension*).

### `if` / `unless`
Estructuras condicionales. `unless` es el opuesto de `if`.

### `import` (Macro)
Trae funciones de un módulo al ámbito actual.

### `in`
Se utiliza en `for` para iterar, o para comprobar pertenencia en una colección.

### `nil`
Representa la ausencia de valor (el átomo `:nil`).

### `not`
Operador de negación booleana estricto.

### `or`
Operador lógico OR booleano estricto.

### `receive`
Bloquea un proceso hasta que recibe un mensaje que coincida con uno de sus patrones.

### `require` (Macro)
Asegura que un módulo esté cargado antes de usar sus macros.

### `super` (Macro)
Permite llamar a la implementación de una función que ha sido sobrescrita (similar al `super` de la OOP).

### `true`
Literal booleano para verdadero (el átomo `:true`).

### `use` (Macro)
Invoca la macro `__using__` de un módulo para inyectar código en el módulo actual.

### `when`
Define guardas en las cláusulas de función o en estructuras de control para añadir condiciones booleanas.

### `with` (Macro)
Encadena operaciones que devuelven tuplas de éxito/error, manejando los fallos de forma elegante.
